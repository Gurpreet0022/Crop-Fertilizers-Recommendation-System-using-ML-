## Crop and Fertilizers Recommendation System using ML

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset 
crop = pd.read_csv('dataset/Crop_recommendation.csv')
```

```python
crop['label'].value_counts()           # Returns the count of each crop(target value) in the dataset
```

```python
# Visualizing the target variable
plt.figure(figsize=(10,5))
sns.countplot(crop['label'])
plt.title('Crop Distribution')
```

```python
# Checking for outliers in the dataset
plt.figure(figsize=(20,12))
crop.boxplot()
plt.title('Checking for Outliers')
```

```python
# Distribution of rainfall
plt.figure(figsize=(20,10))
sns.lineplot(x='label', y='rainfall', data=crop)
plt.title('Distribution of Rainfall')
```

```python
# Relationship between temperature and humidity
plt.figure(figsize=(20,10))
sns.scatterplot(x='temperature', y='humidity', data=crop)
plt.title('Relationship between Temperature and Humidity')
```
